### fpath 점검하고 돌리기

```python
import os 
import numpy as np

fpath = "../../data/synthetic_data/evaluation_data/"

Clean = np.load(fpath + 'clean_data.npy')
# Artifact = np.load(fpath + 'realistic_artifact_data.npy')
Artifact = np.load(fpath + 'increment_artifact_data.npy')

contaminated_by_realistic = Clean + Artifact

np.save(fpath + "contaminated_by_realistic" + ".npy", contaminated_by_realistic)
```

```python
from neurodsp.utils import create_times
import matplotlib.pyplot as plt

# Set some general settings, to be used across all simulations
fs = 2000 #sampling 주파수
n_seconds = 2 #simulation 지속시간
times = create_times(n_seconds, fs) #시간 백터

plt.figure(figsize=(20, 5))

plt.subplot(2, 1, 1)
plt.plot(times, contaminated_by_realistic[0, :], c='xkcd:coral', label = 'Contaminated')
plt.plot(times, Clean[0, :], c='dodgerblue', label = 'Clean')
plt.xlabel("Time (seconds)"); plt.title('contaminated_by_realistic')
plt.legend()
plt.xlim([0, 2])

plt.subplot(2, 1, 2)
plt.plot(times, contaminated_by_increment[0, :], c='xkcd:coral', label = 'Contaminated')
plt.plot(times, Clean[0, :], c='dodgerblue', label = 'Clean')
plt.xlabel("Time (seconds)"); plt.title('contaminated_by_increment')
plt.legend()
plt.xlim([0, 2])

plt.tight_layout()  
```
