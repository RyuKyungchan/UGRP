```python
import numpy as np
import matplotlib.pyplot as plt

# 1.   
fs = 1000  #   (Hz)
t = np.arange(0, 1, 1/fs)  #   (1 )

# 2.  130 Hz  
f_signal = 130  #   (Hz)
signal = np.sin(2 * np.pi * f_signal * t)

# 3. FFT 
N = len(signal)  #  
fft_result = np.fft.fft(signal)
fft_freqs = np.fft.fftfreq(N, 1/fs)  #  

# FFT    (  )
half_N = N // 2
fft_result = np.abs(fft_result[:half_N])
fft_freqs = fft_freqs[:half_N]

# 4.  
plt.figure(figsize=(10, 5))

#    
plt.subplot(1, 2, 1)
plt.plot(t, signal)
plt.title('Time Domain Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

# FFT    
plt.subplot(1, 2, 2)
plt.plot(fft_freqs, fft_result)
plt.title('Frequency Domain Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Parameters for the square wave
sampling_rate = 1000  # Hz
T = 1.0 / sampling_rate  # Sampling interval
L = 1000  # Length of signal
t = np.linspace(0, L*T, L, endpoint=False)  # Time vector

# Generate square wave signal
f = 5  # Frequency of the square wave in Hz
square_wave = np.sign(np.sin(2 * np.pi * f * t))

# Perform FFT
yf = fft(square_wave)
xf = fftfreq(L, T)[:L//2]

# Plot the square wave
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, square_wave)
plt.title('Square Wave')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

# Plot the FFT of the square wave
plt.subplot(2, 1, 2)
plt.stem(xf, 2.0/L * np.abs(yf[:L//2]), 'b', markerfmt=" ", basefmt="-b")
plt.title('FFT of Square Wave')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid()

plt.tight_layout()
plt.show()
```
